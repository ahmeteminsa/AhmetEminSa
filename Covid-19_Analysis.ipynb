## Economic Status vs Total Deaths

let's search there is a relationship between economic status which is created accroding to gdp per capita for countries and total coronavirus deaths

```python
df_13 = df[df['date']<'2021-01-01 00:00:00']
df_13 = df_13 [['continent','country','date','day','month','year','gdp_per_capita','new_deaths','new_cases','new_tests','population']]
df_13 = df_13[df_13['population']>1000000]
df_13 = df_13[df_13['gdp_per_capita']>1]
df_13.head()
```

```python
## df_13['economic_class'] = np.where(df_13['GDP Per Capita'] > 50000, 'Excellent','Bad')
# create a list of our conditions
conditions = [
    (df_13['gdp_per_capita'] <= 5000),
    (df_13['gdp_per_capita'] > 5000) & (df_13['gdp_per_capita'] <= 20000),
    (df_13['gdp_per_capita'] > 20000) & (df_13['gdp_per_capita'] <= 45000),
    (df_13['gdp_per_capita'] > 45000) & (df_13['gdp_per_capita'] <= 70000),
    (df_13['gdp_per_capita'] > 70000)
    ]

# create a list of the values we want to assign for each condition
values = ['Poor', 'Medium-Low', 'Medium', 'Rich','Very Rich']

# create a new column and use np.select to assign values to it using our lists as arguments
df_13['economic_status'] = np.select(conditions, values)

# display updated DataFrame
df_13.head()
```

```python
df_13_eco_status = df_13.groupby(['economic_status','continent','country'])['new_deaths'].aggregate([np.nansum]).reset_index()
df_13_eco_status = df_13_eco_status.rename(columns={'nansum': 'deaths_num'})
df_13_eco_status 
```
